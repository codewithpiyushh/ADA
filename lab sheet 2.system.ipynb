```python
import numpy as np
import matplotlib.pyplot as plt
import time

def makerandomarray(size):
    arr = np.random.randint(1000, size=size)
    return arr

def quicksortavg(arr):
    start_time = time.time()
    quicksort(arr, 0, len(arr) - 1)
    end_time = time.time()
    time_taken = end_time - start_time
    return time_taken

def insertsortavg(arr):
    start_time = time.time()
    insertionsort(arr)
    end_time = time.time()
    time_taken = end_time - start_time
    return time_taken

def swap(arr, i, j):
    temp = arr[i]
    arr[i] = arr[j]
    arr[j] = temp

def partition(arr, l, r):
    pivot = arr[r]
    i = l - 1
    for j in range(l, r):
        if arr[j] < pivot:
            i += 1
            swap(arr, i, j)
    swap(arr, i + 1, r)
    return i + 1

def quicksort(arr, l, r):
    if l < r:
        pi = partition(arr, l, r)
        quicksort(arr, l, pi - 1)
        quicksort(arr, pi + 1, r)

def insertionsort(arr):
    for i in range(1, len(arr)):
        curr = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > curr:
            arr[j + 1] = arr[j]
            j = j - 1
        arr[j + 1] = curr

sizes = [100, 200, 500, 1000, 5000, 10000, 50000, 100000, 500000, 1000000]
timelist = []
for size in sizes:
    arr = makerandomarray(size)
    if size < 500:
        timetaken = insertsortavg(arr)
    else:
        timetaken = quicksortavg(arr)
    timelist.append((size, timetaken))

quicktimelist = []
for size in sizes:
    arr = makerandomarray(size)
    timetaken = quicksortavg(arr)
    quicktimelist.append((size, timetaken))

plt.plot([size for size, _ in timelist], [time for _, time in timelist], label='hybrid')
plt.plot([size for size, _ in quicktimelist], [time for _, time in quicktimelist], label='Quicksort only')
plt.legend()
plt.xlabel('Size of Array')
plt.ylabel('Time taken (seconds)')
plt.title('Comparison Graph')
plt.show()
```
